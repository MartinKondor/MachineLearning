```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

sns.set(font_scale=1.2)
sns.set(rc={'figure.figsize':(9,8)})
```

```python
X = range(-10, 11)

plt.plot(X, [x**2 / 20 for x in X], color='blue')
plt.plot(X, [np.tanh(x) * 5 for x in X], color='orange')
plt.plot(X, [abs(x) for x in X], color='red')

plt.show()
```
