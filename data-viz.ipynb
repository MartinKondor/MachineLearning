```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
plt.hist(np.random.randn(1000))
plt.show()
```

```python
plt.bar(['Red', 'Green', 'Blue'], [5, 3, 2])
plt.show()
```

```python
X = range(-10, 11)

plt.plot(X, [x**2 / 20 for x in X], color='blue')
plt.plot(X, [np.tanh(x) * 5 for x in X], color='orange')
plt.plot(X, [abs(x) for x in X], color='red')

plt.show()
```
